```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset.csv
data = pd.read_csv('32130_AT2_25745973.csv')

# Display the first few rows of the dataset
data.head()
```

```python
data['Age_year'] = (data['age'] / 365).round(3)
data.head()
```

```python
sns.histplot(data['height'], bins=20, kde=True)
plt.title('Height')
plt.xlabel('Height')
plt.ylabel('Frequency')

plt.show()
```

```python
# Calculate quantiles
q25 = np.percentile(data['height'], 25)
q50 = np.percentile(data['height'], 50)
q75 = np.percentile(data['height'], 75)

# Calculate IQR
iqr = q75 - q25
lower_bound = q25 - 2.5 * iqr
upper_bound = q75 + 2.5 * iqr

# Create boxplot
sns.boxplot(x=data['height'])
plt.title('Boxplot of Height')
plt.xlabel('Height')

# Draw quantile lines
plt.axvline(q25, color='blue', linestyle='--', label='25th Quantile')
plt.axvline(q50, color='green', linestyle='--', label='50th Quantile (Median)')
plt.axvline(q75, color='red', linestyle='--', label='75th Quantile')

# Draw outlier bounds
plt.axvline(lower_bound, color='purple', linestyle='-.', label='Lower Bound (Outliers)')
plt.axvline(upper_bound, color='orange', linestyle='-.', label='Upper Bound (Outliers)')

# Add legend
plt.legend()

plt.show()
```

```python
# Plot pie chart of cholesterol
# 1: normal, 2: above normal, 3: well above normal

# Count values for each category
cholesterol_counts = data['cholesterol'].value_counts()
print(data['cholesterol'].value_counts())
# Create labels
labels = ['Normal', 'Above Normal', 'Well Above Normal']

# Custom autopct function to show percentage and count
def autopct_with_count(pct, all_vals):
    absolute = int(round(pct / 100. * sum(all_vals)))
    return f'{pct:.1f}% ({absolute})'

# Plot pie chart
cholesterol_counts.plot.pie(
    autopct=lambda pct: autopct_with_count(pct, cholesterol_counts),
    labels=labels
)
plt.title('Cholesterol')

# Add legend at the right top
plt.legend(labels, title="Cholesterol Levels", bbox_to_anchor=(1, 1))

plt.show()
```
