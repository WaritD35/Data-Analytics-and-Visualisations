```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
import joblib
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('processed_kaggle.csv')
df
```

```python
reserve_id = df['id']
df_classify = df.drop(columns=['id'])
reserve_id
```

```python
df_classify
```

```python
scaler = joblib.load('scaler.pkl')
normalize_columns = ['weight', 'ap_hi', 'ap_lo', 'age_years', 'BMI', 'pulse_pressure']

X_predict[normalize_columns] = scaler.transform(df_classify[normalize_columns])
```

```python
X_predict
```

```python
# Load the model
loaded_mlp_model = joblib.load('mlp_model.pkl')

# ใช้โมเดลที่โหลดไว้ทำนายผล
predictions = loaded_mlp_model.predict(X_predict)
predictions_df = pd.DataFrame(predictions)

#join the predictions with the reserve_id
predictions_df['id'] = reserve_id
predictions_df['cardio'] = predictions_df[0]
predictions_df.drop(columns=[0], inplace=True)
predictions_df

#No: 0 , Yes: 1
predictions_df['cardio'] = predictions_df['cardio'].apply(lambda x: 'Yes' if x == 1 else 'No')

# Save the predictions to a CSV file
predictions_df.to_csv('predictions.csv', index=False)
```
